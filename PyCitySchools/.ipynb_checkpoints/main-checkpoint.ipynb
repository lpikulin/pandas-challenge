```python
import pandas as pd

school_data="Resources/schools_complete.csv"
student_data="Resources/students_complete.csv"

school=pd.read_csv(school_data)

school.head()
```

```python
students=pd.read_csv(student_data)
students.head()
```

```python
#Total Schools
#Total budget
print("The total number of schools is: "+str(school["School ID"].count()))
print("The total budget is: "+str(("${:,.2f}".format(school["budget"].sum()))))
```

```python
combine=pd.merge(school,students,on="school_name",how="outer")
combine.head()
```

```python
print("The total number of students is: "+str("{:,}".format(combine["Student ID"].count())))
print("The average math score is: "+str("{:.2f}".format(combine["math_score"].mean())))
print("The average reading is: "+str("{:.2f}".format(combine["reading_score"].mean())))
```

```python
combine["math_pass"]=combine["math_score"]>=70
print("The % of students passing math is: "+str("{:.2%}".format(combine["math_pass"].mean())))
combine["read_pass"]=combine["reading_score"]>=70
print("The % of students passing reading is: "+str("{:.2%}".format(combine["read_pass"].mean())))
combine["both_pass"]=combine["math_pass"] & combine["read_pass"]
print("The % of students passing reading and math is: "+str("{:.2%}".format(combine["both_pass"].mean())))
```

```python
school_groups =combine.groupby("school_name")

sch_df=pd.DataFrame(
       {"school type":school_groups["type"].first(),
        "students":school_groups["Student ID"].count(),
        "budget":school_groups["budget"].first(),
        "Average Math Score":school_groups["math_score"].mean(),
        "Average Reading Score":school_groups["reading_score"].mean(),
        "Students Passing Math":(school_groups["math_pass"].mean()*100),
        "Students Passing Reading":(school_groups["read_pass"].mean()*100),
        "Students Passing Both":(school_groups["both_pass"].mean()*100)        
       })

        #, budget, stud_budget, avg_math, avg_reading, sch_path_math sch_pass_read, sch_path_both]}
sch_df["Budget per Student"]=sch_df["budget"]/sch_df["students"]
sch_df["Budget per Student"] = sch_df["Budget per Student"].astype(float).map("${:,.2f}".format)
sch_df["budget"]=sch_df["budget"].astype(float).map("${:,.0f}".format)
sch_df["Students Passing Math"]=sch_df["Students Passing Math"].astype(float).map("{:,.2f}%".format)
sch_df["Students Passing Reading"]=sch_df["Students Passing Reading"].astype(float).map("{:,.2f}%".format)
sch_df["Students Passing Both"]=sch_df["Students Passing Both"].astype(float).map("{:,.2f}%".format)
sch_df["Average Math Score"]=sch_df["Average Math Score"].astype(float).map("{:.2f}".format)
sch_df["Average Reading Score"]=sch_df["Average Reading Score"].astype(float).map("{:.2f}".format)
```

```python
sort_df=sch_df.sort_values("Students Passing Both",ascending=False)

sort_df.head(5)

```

```python
grouper=["school_name","grade"]
grade_groups =combine.groupby(grouper)

round(grade_groups["reading_score"].mean(),2)
```

```python
sch_df=pd.DataFrame(
       {"school type":school_groups["type"].first(),
        "students":school_groups["Student ID"].count(),
        "budget":school_groups["budget"].first(),
        "Average Math Score":school_groups["math_score"].mean(),
        "Average Reading Score":school_groups["reading_score"].mean(),
        "Students Passing Math":(school_groups["math_pass"].mean()*100),
        "Students Passing Reading":(school_groups["read_pass"].mean()*100),
        "Students Passing Both":(school_groups["both_pass"].mean()*100)        
       })

        #, budget, stud_budget, avg_math, avg_reading, sch_path_math sch_pass_read, sch_path_both]}
sch_df["Budget per Student"]=sch_df["budget"]/sch_df["students"]
sch_df["Budget per Student"].quantile([.25,.5,.75])
```

```python
bins=[0,600,625,650,10000]
group_labels=["under 600","600 to <625","625 to <650","650 and over"]
sch_df["Spend Group"] = pd.cut(sch_df["Budget per Student"], bins, labels=group_labels)

spend_group = sch_df.groupby("Spend Group")

spend_df=pd.DataFrame(
       {"Average Math Score":spend_group["Average Math Score"].mean(),
        "Average Reading Score":spend_group["Average Reading Score"].mean(),
        "Students Passing Math":spend_group["Students Passing Math"].mean(),
        "Students Passing Reading":spend_group["Students Passing Reading"].mean(),
        "Students Passing Both":spend_group["Students Passing Both"].mean()        
       })
spend_df["Students Passing Math"]=spend_df["Students Passing Math"].astype(float).map("{:,.2f}%".format)
spend_df["Students Passing Reading"]=spend_df["Students Passing Reading"].astype(float).map("{:,.2f}%".format)
spend_df["Students Passing Both"]=spend_df["Students Passing Both"].astype(float).map("{:,.2f}%".format)
spend_df["Average Math Score"]=spend_df["Average Math Score"].astype(float).map("{:.2f}".format)
spend_df["Average Reading Score"]=spend_df["Average Reading Score"].astype(float).map("{:.2f}".format)

spend_df
```
